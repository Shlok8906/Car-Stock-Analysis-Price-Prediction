#Predicting Car Prices

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# ---- FEATURES ----
features = [
    'Year', 'Engine', 'Max Power', 'Kilometer',
    'Fuel Tank Capacity', 'Width', 'Height',
    'Fuel Type', 'Transmission', 'Owner'
]

df = cleaned_df.copy()
df = df.dropna(subset=['Price'])
df = df[(df['Price'] > 50000) & (df['Price'] < 4000000)]   # remove outliers

X = df[features]
X = pd.get_dummies(X, drop_first=True)
y = df['Price']

# ---- LOG TRANSFORM ----
y_log = np.log1p(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_log, test_size=0.2, random_state=42
)

# ---- MODEL ----
model = RandomForestRegressor(
    n_estimators=500,
    max_depth=25,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

model.fit(X_train, y_train)

# ---- PREDICT ----
y_pred_log = model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_true = np.expm1(y_test)

# ---- METRICS ----
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f"R Score: {r2*100:.2f}%")
print(f"RMSE: {rmse:,.2f}")
print(f"MAE: {mae:,.2f}")
```

```python
import numpy as np
import pandas as pd

def predict_car_price():
    print("Enter the car details below:\n")

    year = float(input("Year: "))
    engine = float(input("Engine (in CC): "))
    max_power = float(input("Max Power (in bhp): "))
    km = float(input("Kilometer Driven: "))
    fuel_tank = float(input("Fuel Tank Capacity: "))
    width = float(input("Width (mm): "))
    height = float(input("Height (mm): "))
    fuel_type = input("Fuel Type (Petrol/Diesel/CNG): ")
    transmission = input("Transmission (Manual/Automatic): ")
    owner = input("Owner (First/Second/Third): ")

    # ----- Prepare input in same format as training -----
    data = pd.DataFrame({
        'Year': [year],
        'Engine': [engine],
        'Max Power': [max_power],
        'Kilometer': [km],
        'Fuel Tank Capacity': [fuel_tank],
        'width': [width],
        'height': [height],
        'Fuel Type': [fuel_type],
        'Transmission': [transmission],
        'Owner': [owner]
    })

    # match training encoding
    data = pd.get_dummies(data)
    X_cols = X.columns   # Columns used during training
    data = data.reindex(columns=X_cols, fill_value=0)

    # ----- Predict (model predicts log price) -----
    pred_log = model.predict(data)[0]

    # convert back to normal price
    pred_price = np.expm1(pred_log)

    print("\n--------------------------------")
    print(f"Predicted Car Price:  {pred_price:,.2f}")
    print("--------------------------------")


predict_car_price()
```
